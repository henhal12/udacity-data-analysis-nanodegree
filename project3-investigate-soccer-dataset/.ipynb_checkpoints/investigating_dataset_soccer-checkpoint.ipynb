### Exporting Dataset Using SQL Queries

>The datasets are exported to csv files using DB Browser for SQLite. 
>
>For the "match_full" dataset, I join the Country table, Team table, and Matches table to show all the necessary information. The query is given below:
>
>>SELECT m.id, l.name AS league_name, c.name AS country_name,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; m.season, m.stage, m.date, ht.team_long_name AS home_team,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; at.team_long_name AS away_team, m.home_team_goal AS home_goal,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; m.away_team_goal AS team_goal<br>
JOIN League l ON m.league_id = l.id<br>
JOIN Country c ON m.country_id = c.id<br>
JOIN Team ht ON m.home_team_api_id = ht.team_api_id<br>
JOIN Team at ON m.away_team_api_id = at.team_api_id<br>
WHERE country_name IN ('England', 'Italy', 'Spain', 'Germany')<br>
GROUP BY 1<br>
ORDER BY 4, 5<br>

>For the "team_attribute" dataset, I join the Team table and Team Attribute table using this query:
>
>>SELECT t.team_api_id, t.team_long_name AS team_name, ta.* <br>
FROM Team t <br>
JOIN Team_Attributes ta ON ta.team_api_id = t.team_api_id <br>
ORDER BY 1<br>

>For the "player_attribute" dataset, I join the Player table with Player Attribute table using this query:
>
>>SELECT * <br>
FROM Player p <br>
JOIN Player_Attributes pa ON p.player_api_id = pa.player_api_id <br>

### What team has improved the most over the time period?

> In order to answer the question, I have to make new columns to count which team won and lost. The code below will do the job.

```python
df_match.loc[df_match['home_goal'] > df_match['team_goal'], 'win_team'] = df_match['home_team']
df_match.loc[df_match['home_goal'] > df_match['team_goal'], 'lose_team'] = df_match['away_team']
df_match.loc[df_match['home_goal'] < df_match['team_goal'], 'win_team'] = df_match['away_team']
df_match.loc[df_match['home_goal'] < df_match['team_goal'], 'lose_team'] = df_match['home_team']

df_match.head()
```

> Then, I count the total lost match for each team and find the lose percentage by deviding the lose count to the total match.

```python
df_lose = df_match.groupby(['season','country_name'])['lose_team'].value_counts().reset_index(name = 'lose_count')
df_lose['lose_percentage'] = df_lose['lose_count'] / sum(df_lose['lose_count']) * 100

df_lose.head(10)
```

> I rename the "lose_team" column label to "team_name" for neatness.

```python
df_lose.rename(columns={'lose_team': 'team_name'}, inplace=True)

df_lose.head(10)
```

> I make a pivot table to put the team name in each column. The row represent the lose percentage in each season.

```python
lose_pivot_all = df_lose.pivot(index='season', columns = 'team_name', values='lose_percentage')

lose_pivot_all
```

> As we can see, there are a lot of teams that have null values in the data table. It is because some of them are not playing all eight seasons in their league. For making better comparison, I decide to drop all the teams that do not play in all eight seasons.

```python
lose_pivot_all.dropna(axis= 1, how='any', inplace=True)

lose_pivot_all
```

> Next, I find the lose percentage difference for each season and take the sum.

```python
lose_sumdiff_all = lose_pivot_all.sub(lose_pivot_all.shift()).sum().sort_values()

lose_sumdiff_all
```

> Finally, I plot the sum of differences for each team and analyze the result.

```python
ax = lose_sumdiff_all.plot(kind='bar',figsize=(15,12), colormap='jet')
plt.title('Sum of Lose Percentage Difference for Top Teams in All Four Leagues', fontsize= 20)
plt.xlabel(''); plt.ylabel('Sum of Lose Percentage Difference',fontsize=15)
plt.xticks(fontsize=12); plt.yticks(fontsize=12)
ax.set_facecolor('lightgrey')
```

# Question 1 Answer:

> The plot above represent thelose percentage sum of difference for each team. It goes from negative number to positive number. The negative number means the team improved from season to season. The improvement means that the lose percentage will decrease in each season resulting in negative number for the sum.<br>
<br>
> <b> Napoli from Italy shows the most improvement across all season followed by Tottenham Hotspur and Machester City from England. </b>

### What team attributes lead to the most victories?

> As I did in the previous section, I count the number of wins each team has for each season. Then, I calculate their winning percentage for each season.

```python
df_win = df_match.groupby('season')['win_team'].value_counts().reset_index(name = 'win_count')
df_win['win_percentage'] = df_win['win_count'] / sum(df_win['win_count']) * 100

df_win.head(10)
```

> I rename the "win_team" label to "team_name".

```python
df_win.rename(columns={'win_team': 'team_name'}, inplace=True)

df_win.head(10)
```

> I create a pivot table to make the winning percentage is easier to analyze.

```python
win_pivot = df_win.reset_index().pivot(index='season', columns = 'team_name', values='win_percentage')

win_pivot
```

> When we analyzed the lose percentage in question one, we saw null values because many of the teams did not play in all eight seasons. In this data table, we also experience the same problem. Consequently, I delete all the teams that do not appear in all seasons.

```python
win_pivot.dropna(axis=1,inplace=True)

win_pivot
```

> I average the winning percentage for these teams and sort them from highest to lowest.

```python
win_pivot_mean = win_pivot.mean().sort_values(ascending = False)

win_pivot_mean
```

> I present the win percentage average in a bar plot for analysis.

```python
ax = win_pivot_mean.plot(kind='bar', figsize=(15,12), colormap='jet')
plt.title('Average Win Percentage Across All Four Leagues', fontsize= 20)
plt.xlabel(''); plt.ylabel('Average Win Percentage', fontsize=15)
plt.xticks(fontsize=12); plt.yticks(fontsize=12)
ax.set_facecolor('lightgrey')
```

> <b> From the figure, it appears that FC Barcelona from Spain League has the highest winning percentage across all seasons.</b><br>
<br>

### Does older player has higher overall rating than younger player?

> To answer the question, I create a function that change the data type for "date" table to datetime object and use it to find the player age. The recent date that I use is 2016-01-01 because the match data goes until year of 2016.

```python
def get_age(x):
    date = x.split(" ")[0]
    born = dt.datetime.strptime(date, "%Y-%m-%d").date()
    recent_date = dt.datetime.strptime("2016-01-01", "%Y-%m-%d").date()
    return recent_date.year - born.year - ((recent_date.month, recent_date.day) < (born.month, born.day))

df_player_att['age'] = df_player_att['birthday'].apply(lambda x:get_age(x))
```

> The player attribute dataframe shows that each player has more than one data. So, I take the average of the data for each player.

```python
df_player_att = df_player_att.groupby('player_name').mean().round()
```

> Then, I plot the histogram for the age to see the distribution.

```python
df_player_att['age'].hist(figsize=(7,5), color='navy')
plt.title('Data Distribution By Age', fontsize=13)
plt.xlabel('Age', fontsize=13)
```

> The distribution is skewed to the right and we can see that there are some outliers in the dataset. So, I make a boxplot to see which are the outliers.

```python
plt.figure(figsize=(7,7))
plt.boxplot(df_player_att['age'])
plt.title('Age Boxplot', fontsize=13)
plt.ylabel('Age',fontsize=13)
```

> The boxplot shows most of the data concentrated between age 24 to 32. Then, I query the dataframe to drop the age outside the boxplot.

```python
new_df_player_att = df_player_att.query('age >=24 and age <= 32')
```

> I find the "overall rating" column statistic using the code below.

```python
new_df_player_att['overall_rating'].describe()
```

> Next, I categorize the overall rating to three groups: low, medium, and high, based on the given statistic. Then, I assign the category using this function.

```python
def get_rating_class(x):
    if x <= 64:
        return 'low'
    elif x >= 64 and x <= 71:
        return 'mid'
    else:
        return 'high'

new_df_player_att['rating_class'] = new_df_player_att['overall_rating'].apply(lambda x:get_rating_class(x))
```

> I create three separate table based on the categories for easier analysis.

```python
low = new_df_player_att.query('rating_class == "low"')
mid = new_df_player_att.query('rating_class == "mid"')
high = new_df_player_att.query('rating_class == "high"')
```

> Finally, I find the average age for each category and plot it into a bar chart.

```python
low_age_mean = low['age'].mean()
mid_age_mean = mid['age'].mean()
high_age_mean = high['age'].mean()
```

```python
x = [1, 2, 3]
y = [low_age_mean, mid_age_mean, high_age_mean]

plt.figure(figsize=(7,7))
plt.bar(x, y, color='navy')
plt.title('Average Age By Overall Rating Class',fontsize=13)
plt.xticks((1,2,3),('Low','Mid','High'), fontsize=13)
plt.yticks(np.arange(0, 30, step=3))
plt.ylabel('Average Age',fontsize=13)
```

# Question 3 Answer:

> <b>The result shows that as the overall rating class increase, the age is also increasing. It confirms that older player has higher overall rating. I think it is because at older age, the player has more experience and also better skills, thus their overall rating are higher. </b>
